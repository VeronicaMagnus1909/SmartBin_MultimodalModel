### Convert to Meltrospecgram for one file audio

```python
y, sr = librosa.load(audio_files_GLASS[0], sr= 22050)

sgram = librosa.stft(y)

sgram_mag, _ = librosa.magphase(sgram)


# Generate Mel spectrogram
mel_scale_sgram = librosa.feature.melspectrogram(S = sgram_mag,sr = sr)

# Convert to dB
mel_scale_sgram_db = librosa.amplitude_to_db(mel_scale_sgram, ref=np.min)

# Save Mel spectrogram as image
#plt.figure(figsize=(10, 10))
mel_sgram = librosa.amplitude_to_db(mel_scale_sgram_db, ref=np.min)
librosa.display.specshow(mel_sgram, sr= sr, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
```
